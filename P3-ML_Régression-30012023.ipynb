# Outils d'évaluation des algorithmes.

1) Calculer de la distance entre valeurs prédites et vraies valeurs: 
- La somme des carrés des résidus (RSS).

- La moyenne de cette somme (MSE).

- La racine carrée de cette moyenne (RMSE).  

2) La corrélation entre valeurs prédites et vraies valeurs :  

- Le coefficient de détermination : (R2)

# Modèle utilisé

- Régression linéaire :  
L'elastic net combine les normes ℓ1 et ℓ2 pour obtenir une solution moins parcimonieuse que le lasso, mais plus stable et dans laquelle toutes les variables corrélées pertinentes pour la prédiction de l'étiquette sont sélectionnées et reçoivent un poids identique.


## Df 1 : sans ENERGYSTARScore

### Target :  SiteEUIWN(kBtu/sf)

#### CrossValidation

```python
# Set our Elastic Model to 6 splits with a Shuffle before and a aleatory based on 5.
kf = KFold(n_splits=6, shuffle=True, random_state=5)
```

```python
# Create our Elastic Model
Elastic = ElasticNet(random_state=0)
```

```python
cv_scores = cross_val_score(Elactic, X_Df1_Energie_train, y_Df1_Energie_train, scoring="neg_mean_squared_error", cv=kf)
```

```python
print(cv_scores)
```

```python
rmse = np.sqrt(-cv_scores)
print("Average RMSE: {}".format(np.mean(rmse)))

# Print the standard deviation
print(np.std(cv_scores))

# Print the 95% confidence interval
print(np.quantile(cv_scores, [0.025, 0.975]))
```

#### GridSearchCV

```python
# Set up the parameter grid
param_grid = {'alpha': np.linspace(0.00001, 1, 10), # We get 10 values
              'l1_ratio' : np.arange(0.0, 1, 0.1)} # We incremente by 0.1
```

```python
Elastic_cv = GridSearchCV(Elactic, param_grid, cv=kf)
```

```python
Elastic_cv.fit(X_Df1_Energie_train, y_Df1_Energie_train)
```

```python
y_pred = Elastic_cv.predict(X_Df1_Energie_test)
```

```python
print("Elastic_cv Regression Parameters: {}, R2: {}".format(Elastic_cv.best_params_, Elastic_cv.score(X_Df1_Energie_test, y_Df1_Energie_test)))
```

#### Evaluation du modèle 

```python
# R2
r_squared = Elastic_cv.score(X_Df1_Energie_test, y_Df1_Energie_test)

# MSE
median_squared_error = median_absolute_error(y_Df1_Energie_test, y_pred)

# Compute RMSE
rmse = mean_squared_error(y_Df1_Energie_test, y_pred, squared=False)

#Time
start_time = time.time()
Elastic_cv.fit(X_Df1_Energie_train, y_Df1_Energie_train)
end_time = time.time()
time_taken = end_time - start_time

# Print the metrics
print("R^2: {}".format(r_squared))
print("RMSE: {}".format(rmse))
print("Median Squared Error: {}".format(median_squared_error))
print("Time taken for model to run: {:.2f} seconds".format(time_taken))
```
