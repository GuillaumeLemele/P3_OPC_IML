# Définitions des fonctions

```python
def model_performance(model, nom_model, X_train, y_train, X_test, y_test, target, energy_star_score='Oui'):
           
    
    # Mesure du temps d'entraînement 
    start_time = time.time()
    
    # Entraînement du modèle  
    model.fit(X_train, y_train)
    end_time = time.time()
    time_taken = end_time - start_time
    
    # Prédictions
    y_pred = model.predict(X_test)
    
    # Meilleurs paramètres
    best_params = model.best_params_
    
    # Evaluation du modèle
    mae = mean_absolute_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    median_squared_error = median_absolute_error(y_test, y_pred)
    r2 = metrics.r2_score(y_test, y_pred)
    
    # Df des résultats
    
    results = {
        # Renseignement du modèle dans le dataset
        "Modèle" : nom_model,
        # Renseignement de la target dans le dataset
        "Target" : target,
        # Renseignement de la présence de cette variable dans le dataset
        "EnergyStar_Score_Var." : energy_star_score,
        "Best Parameters": str(best_params),
        "MAE": mae,
        "RMSE": rmse,
        "Median Squared Error": median_squared_error,
        "R^2": r2,
        "Time taken": time_taken}

    df = pd.DataFrame(data=results, index=['Modèle'])
    display(df)
    
    # Plot des résultats 
    
    err_hist = np.abs(y_test - y_pred)
    plt.figure(figsize=(8,4))
    plt.title('')
    plt.xlabel('')
    plt.ylabel('Nombres d erreurs de prédiction', fontweight='bold')
    plt.hist(err_hist, bins = 55);
```

# Outils d'évaluation des algorithmes.

- Séparation d'un échantillon d'entraînement d'avec un échantillon test (avec fixation de l'aléatoire, et nous le mélangeons avant le split).


- Mise en place d'une validation croisée (Cross Validation) : avec un nouvel intermédiaire : le validation set, afin de garder l'échantillon du test set pour l'évaluation finale, et ainsi éviter les risques de biais sur le test set (car un mélange aléatoire lors d'une séparation train test peux induire un biais malgré tout sur le test set, l'ajout d'un intermédiaire renforce le pouvoir de l'aléatoire). 


- Nous avons le choix entre un KFold, un ShuffleSplit, un stratifiedKFold, ou un GroupKFold :  
KFold divise les données d'entraînement en plusieurs plis, où chaque pli contient un train set et un validation set (qui normalement serait notre test set); différent sur le même échantillon.  
ShuffleSplit : Mélange aléatoire avec les même proportions pour les train et validations set sur le nombre de plis choisis.
StratifiedKFold : similaire à KFold mais assure une distribution équitable des classes cibles dans chaque pli, ce qui est important lorsque les données sont déséquilibrées.  
GroupKFold : Utile sur un dataset où les données dépendent de certains groupes. Ces groupes représentent un facteur d'influence dans notre dataset. Similaire à KFold mais ajoute la contrainte de conserver les éléments appartenant au même groupe dans le même pli.



- Nous choisissons un KFolfd, car dans un modèle d'apprentissage automatisé de régression, cette technique convient très bien car nous n'avons pas de contrainte de classe sur nos données; avec un nombre de fold égal à 5 et une fixation de l'aléatoire égale à 5.


- GridsearchCV : cet outils nous permet de trouver le modèle avec les meilleurs hyperparamètres en comparant les différentes performances de chaque combinaison en utilisant l'outils de Cross Validation. Nous enregistrons les hyperparamètres que nous voulons tester avec leurs valeurs asoociés dans un dictionnaire. Enfin, la grille fonctionnant comme un estimateur, nous faisons passer le modèle choisi, le dictionnaire crée, et notre Cross Validation au sein de notre objet GridsearchCV. 


# Mesures d'évaluation des algorithmes.

1) Nous commençons par calculer la distance entre nos valeurs prédites et les vraies valeurs :   
  
  - La moyenne des erreurs absolues (MAE) : l'importance d'une erreur est linéaire avec son amplitude.  
Meilleur évaluation de la performance du modèle / à la RMSE si présence de valeurs aberrantes.  
    
    
  - La racine carré de la moyenne quadratique des erreurs (RMSE) : l'importance d'une erreur est exponentielle avec son amplitude.  
Elle pénalise donc beaucoup plus les grandes erreurs que la MAE  
Exemple : Une erreur de 10 mètre devient 100 fois plus grave qu'une erreur de 1 mètre.  
    
    
  - La médiane des erreurs absolues (Median_Abs_Error) : tri les erreurs dans l'ordre pour en sortir la médiane.  
Très utile pour évaluer la performance du modèle sans prendre en compte les valeurs aberrantes par exemple.

2) La corrélation entre valeurs prédites et vraies valeurs :  

- Le coefficient de détermination (R2) : Sklearn implémenter dans la méthode score de base pour les régressions.  
Evalue la performance du modèle par rapport au niveau de variation présent dans les données.  
R2 = 1 - Erreur/Variance. Score entre 0 et 1.  
Proche de 1, les erreurs de notre modèle sont beaucoup plus petite que la variance présente dans nos données.

# Modèles utilisés

Nous allons utiliser plusieurs modèles pour tenter d'atteindre nos objectifs de prédiction avec les modèles les plus performants selon l'intensité carbone ou la consommation d'énergie normalisée.  
Nous allons pour cela utiliser des algorithmes qui ont des stratégies différentes pour atteindre le même objetif :  
ElasticNet, Random Forest et Support Vector Machines (SVM), qui sont souvent utilisés dans la communauté des Data Scientist.

- ElasticNet :
C'est un algorithme de régression dit **linéaire** qui utilise une combinaison linéaire des régressions Ridge et Lasso.  
Il est utile pour traiter les cas où les variables sont colinéaires entres elles, car l'elastic net combine les normes ℓ1 et ℓ2 pour obtenir une solution moins parcimonieuse que le lasso, mais plus stable et dans laquelle toutes les variables corrélées pertinentes pour la prédiction sont sélectionnées et reçoivent un poids identique.  
Les hyperparamètres importants pour nous sont :  
Alpha : il contrôle la force de la régularisation.  
l1_ratio : ici la particularité de l'ElasticNet par rapport à un simple Ridge ou Lasso : cet hyperparamètre contrôle le mélange entre la régularisation Ridge et Lasso. Par exemple : si le L1 ratio est égal à 1, notre modèle équivaut à un Lasso, s'il est égal à 0, notre modèle équivaut à un Ridge.  
    
      
      
- Random Forest :
C'est un algorithme de régression/classification dit **ensembliste**.  
L'ensemble learning est une technique qui entraîne plusieurs modèles d'apprentissage automatisés en parallèle pour ensuite considérés l'ensemble de leurs prédictions (Loi des grands nombres ; avec au moins 51% de performances pour converger vers les 100%, un certains nombre de modèles qui plus est comportant une diversité minimale au sein des différents modèles).  
L'algorithme des forêts aléatoires se base sur la méthode de bagging (Bootstrapping; échantillonage aléatoire en replaçant les données après tirage au sort, soit plus communément appelé 'échantillonage aléatoire avec remplacement').  
Nous créons donc un ensemble d'arbres de décisions (par défault) sur la base de données tirées au hasard pour les regrouper et faire ensuite notre prédiction (ce qui réduit la variance de nos premiers modèles).  
Les hyperparamètres importants pour Random Forest incluent :  
n_estimators : nombre d'arbres dans la forêt aléatoire.  
max_depth : profondeur maximale des arbres de décision.  
  
  
- Il est aussi possible d'utiliser d'autres modèle comme AdaBoost et GradientBoosting qui utilisent une technique tel que le boosting (entrainement l'un après l'autre plusieurs modèles faibles qui se corrigent en série pour devenir plus fort afin de réduire 'le biais de la foule').  

- Il est aussi possible d'utiliser d'autres modèle  qui utilisent la technique du stacking (Entrainement d'un modèle d'apprentissage automatisé par dessus les prédictions de notre foule au lieu de rassembler les prédictions de notre foule pour retenir une prédiction majoritaire. Ici notre modèle supérieur apprend à reconnaitre qui a tord et à raison dans notre foule pour prédire les résultats.).


- Support Vector Machine  (Ici Support Vector Regression):
C'est un algorithme de régression/classification qui a pour but est de trouver un hyperplan qui minimise la quantité d'erreur entre les prédictions et les valeurs cibles réelles. L'algorithme minimise la somme des erreurs en travaillant avec des points de données qui sont définis comme des "vecteurs de support".  
Les hyperparamètres importants pour le SVM incluent :  
C : contrôle la force de la régularisation. Elle contrôle la marge entre les prédictions et les valeurs cibles réelles. Une valeur plus petite de C permet une marge plus grande et une réduction de la variance au détriment de la performance, alors qu'une valeur plus grande augmente la performance au détriment de la variance..
kernel : type de noyau utilisé pour effectuer la transformation des données. C'est une fonction qui transforme les données d'entrée en un nouvel espace. Plusieurs choix sont possibles tels que la fonction radiale (RBF), la fonction polynomiale et la fonction linéaire.   
gamma : contrôle la complexité de la fonction de transformation. Une valeur plus petite de gamma augmente la variance et une valeur plus grande réduit la variance.
